```python
data=pd.read_csv('Exam_Score_Prediction.csv')
data.head()
```

```python
data=data.drop('student_id')
```

```python
categorical_col=['gender', 'course', 'internet_access', 'sleep_quality', 'study_method', 'facility_rating', 'exam_difficulty' ]
data_enc=pd.get_dummies(data,columns=categorical_col, drop_first=True)
```

```python
X=data_enc.drop('exam_score', axis=1)
y=data_enc['exam_score']
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2)
```

**Лучшие параметры: {'max_depth': 7, 'min_samples_leaf': 1, 'min_samples_split': 10}**

```python
model=DecisionTreeRegressor(   random_state=42, 
    max_depth=7,   
    min_samples_split=10,
    min_samples_leaf=1)
```

```python
model.fit(X_temp, y_temp)
```

```python
pred=model.predict(X_test)
```

```python
print('--'*15)
print(f'(MSE):')
print(mean_squared_error (y_test, pred))
print('--'*15)
print(f'(MAE):')
print(mean_absolute_percentage_error(y_test, pred))
print('--'*15)
print(f'(R2_score):')
print(r2_score (y_test, pred))
print('--'*15)
```

```python
from sklearn. model_selection import GridSearchCV
```

```python
param_grid={
    'max_depth':[3, 4, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 3, 5]
    }
grid=GridSearchCV(
    DecisionTreeRegressor(random_state=42),
    param_grid,
    cv=5,
    scoring='neg_mean_squared_error')
grid.fit(X_temp, y_temp)
print('***'*30)
print("Лучшие параметры:", grid.best_params_)
best_model = grid.best_estimator_
print('***'*30)
```
