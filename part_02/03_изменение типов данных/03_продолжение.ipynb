**Обработка ошибок: try-except**

```python

position = [
    ['2data9-05-data', '6'],
    ['2data9-05-02', '5'],
    ['2data9-05-03', '5'],
    ['2data9-05-04', '4'],
    ['2data9-05-05', '5'],
    ['2data9-05-06', '5'],
    ['2data9-05-07', '4'],
    ['2data9-05-08', '4'],
    ['2data9-05-09', '3'],
    ['2data9-05-10', '3'],
]
```

```python
count_lines=0
for i in position:
    count_lines+=1
print(count_lines)
```

```python
total=0
for i in position:
    level = int(i[1])
    total+=level
print(total)
    
```

```python
print(total/count_lines)
```

```python

position_1 = [
    ['2019-05-01', '- 6'],
    ['2019-05-02', '+5'],
    ['2019-05-03', ' 5'],
    ['2019-05-data', '4'],
    ['2019-05-05', '5'],
    ['2019-05-06', '5'],
    ['2019-05-07', '4'],
    ['2019-05-08', 'Error 5'],
    ['2019-05-09', '3'],
    ['2019-05-10', '3'],
]
```

```python
data=pd.read_excel('seo_data.xlsx', sheet_name='traffic_data', engine='openpyxl')
subcategory_dict =pd.read_excel('seo_data.xlsx', sheet_name='subcategory_ids', engine='openpyxl')
```

```python
data.head()
```

```python
subcategory_dict.head()
```

```python
data_final =data.merge(subcategory_dict, on='subcategory_id', how='left' )
```

```python
data_final.head()
```

```python
category_dict=pd.read_excel('seo_data.xlsx', sheet_name='category_ids')
category_dict.head()
```

```python
data_final.merge(category_dict, on='category_id', how='left')
```

**Сводные таблицы**

```python
data_final_2=pd.read_csv('data_final.csv')
data_final_2.head()
```

```python
data_pivot =data_final.pivot_table(index=['category_name', 'subcategory_name'], columns='source', values='visits', aggfunc='sum')
```

```python
data_pivot.head()
```

```python
data_pivot['ratio']=data_pivot['organic']/data_pivot['direct']
```

```python
data_pivot.head()
```
